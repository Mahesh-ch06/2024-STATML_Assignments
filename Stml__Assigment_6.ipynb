**Question 1:**

Imagine you are developing a simple machine learning model to predict whether a student will pass a course based on the following five features:

Study Hours per Week: Number of hours the student studies per week.

Class Attendance (%): Percentage of classes attended by the student.

Previous Exam Score (%): Score in the previous exam.

Participation in Group Study: Whether the student participates in group study sessions (1 for Yes, 0 for No).

Submission of Assignments (%): Percentage of assignments submitted by the student.

Dataset

Below is an example dataset with 10 students (rows), 5 features, and a binary target label (Pass/Fail). The target label is 1 if the student is predicted to pass, and 0 if they are predicted to fail.

**Dataset**

```python
data = [
    [1, 10, 90, 85, 1, 100, 1],
    [1, 5, 60, 70, 0, 80, 0],
    [1, 15, 95, 90, 1, 100, 1],
    [1, 2, 30, 50, 0, 40, 0],
    [1, 12, 85, 88, 1, 90, 1],
    [1, 8, 75, 65, 0, 70, 0],
    [1, 14, 92, 93, 1, 95, 1],
    [1, 3, 50, 55, 0, 60, 0],
    [1, 11, 88, 80, 1, 85, 1],
    [1, 6, 65, 60, 0, 75, 0]
]
```

**Number of Epochs**

```python
X = [row[:-1] for row in data]
y = [row[-1] for row in data]

learning_rate = 0.001
epochs = 10

def activation_function(z):
    return 1 if z >= 0 else 0

def perceptron(X, y, weights, learning_rate, epochs):
    x0 = 1
    w0, w1, w2, w3, w4, w5 = weights

    for epoch in range(epochs):
        correct_predictions = 0
        y_pred = [0] * len(y)

        for i in range(len(X)):
            z = (w0 * X[i][0]) + (w1 * X[i][1]) + (w2 * X[i][2]) + (w3 * X[i][3]) + (w4 * X[i][4]) + (w5 * X[i][5])
            y_pred[i] = activation_function(z)

            if y_pred[i] != y[i]:
                w0 += learning_rate * (y[i] - y_pred[i]) * X[i][0]
                w1 += learning_rate * (y[i] - y_pred[i]) * X[i][1]
                w2 += learning_rate * (y[i] - y_pred[i]) * X[i][2]
                w3 += learning_rate * (y[i] - y_pred[i]) * X[i][3]
                w4 += learning_rate * (y[i] - y_pred[i]) * X[i][4]
                w5 += learning_rate * (y[i] - y_pred[i]) * X[i][5]
            else:
                correct_predictions += 1

        accuracy = (correct_predictions / len(y)) * 100
        print(f"Epoch {epoch + 1}/{epochs} - Accuracy: {accuracy:.2f}%")
        print(f"Actual y: {y}")
        print(f"Predicted y: {y_pred}")
        print()

    return [w0, w1, w2, w3, w4, w5]
```

**Weight Initialization**

```python
initial_weights = [1, 1, 1, 1, 1, 1]
final_weights = perceptron(X, y, initial_weights, learning_rate, epochs)

print(f"Final Weights: {final_weights}")
```

```python
features = [
    "Bias Term",
    "Study Hours per Week",
    "Class Attendance (%)",
    "Previous Exam Score (%)",
    "Participation in Group Study",
    "Submission of Assignments (%)"
]
for feature, weight in zip(features, final_weights):
    print(f"{feature}: {weight:.6f}")
```

Implement a perceptron learning model to train above data, and find the accuracy.

```python
import numpy as np
import pandas as pd
data = pd.read_csv('/content/diabetes_data_upload[1].csv')
data
```

**Initialize Parameters**

```python
x0 = 1
x1 = data['Age']
x2 = data['Gender'].map({'Male': 1, 'Female': 0})
x3 = data['Polyuria'].map({'Yes': 1, 'No': 0})
x4 = data['Polydipsia'].map({'Yes': 1, 'No': 0})
x5 = data['sudden weight loss'].map({'Yes': 1, 'No': 0})
x6 = data['weakness'].map({'Yes': 1, 'No': 0})
x7 = data['Polyphagia'].map({'Yes': 1, 'No': 0})
x8 = data['Genital thrush'].map({'Yes': 1, 'No': 0})
x9 = data['visual blurring'].map({'Yes': 1, 'No': 0})
x10 = data['Itching'].map({'Yes': 1, 'No': 0})
x11 = data['Irritability'].map({'Yes': 1, 'No': 0})
x12 = data['delayed healing'].map({'Yes': 1, 'No': 0})
x13 = data['partial paresis'].map({'Yes': 1, 'No': 0})
x14 = data['muscle stiffness'].map({'Yes': 1, 'No': 0})
x15 = data['Alopecia'].map({'Yes': 1, 'No': 0})
x16 = data['Obesity'].map({'Yes': 1, 'No': 0})

y = data['class'].map({'Positive': 1, 'Negative': 0})
```

**The Perceptron**

```python
x = np.array(list(zip(x1, x2, x3, x4, x5, x6, x7, x8, x9, x10, x11, x12, x13, x14, x15, x16)))
y = np.array(y)

def initialize_weights(n):
    return np.random.randn(n)

def activation_function(z):
    return 1 if z > 0 else 0

def perceptron(x, y, weights, learning_rate, target_accuracy):
    n_samples, n_features = x.shape
    weights = np.random.randn(n_features + 1)
    x = np.hstack([np.ones((n_samples, 1)), x])

    epochs = 0
    while True:
        y_pred = np.array([activation_function(np.dot(x[i], weights)) for i in range(n_samples)])
        errors = y - y_pred

        if np.all(errors == 0):
            break

        for i in range(n_samples):
            if errors[i] != 0:
                weights += learning_rate * errors[i] * x[i]

        accuracy = np.mean(y_pred == y) * 100
        epochs += 1

        if accuracy >= target_accuracy:
            print(f"Epoch: {epochs}")
            print(f"Final Accuracy: {accuracy:.2f}%")
            break

    return weights

learning_rate = 0.01
target_accuracy = 80

weights = perceptron(x, y, initialize_weights(x.shape[1]), learning_rate, target_accuracy)
print("\nFinal weights:", weights)
```
